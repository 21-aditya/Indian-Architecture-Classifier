```python
import os
import csv
```

```python
features = ['Mughal','Rajput','Chola','Hoysala','Kalinga','Tibetan','Kerala','MaruGurjara','Cave-Arch']
for item in list:
  data = []
  idx = 1
  for feature in features:
    cur_images = os.listdir(f'{item}/{feature}')
    for image in cur_images:
      x = []
      x.append(idx)
      x.append(feature + '/' + image)
      x.append(feature)
      idx = idx + 1
      data.append(x)
  header = ['Sno', 'Image', 'Class']
  with open(f'{item}.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(data)
```
